#**Finding out the number of toss wins w.r.t each team**

```python
toss = df['toss_winner'].value_counts()
toss
```

```python
toss_teams = list(df['toss_winner'].value_counts().keys())
toss_teams
```

```python
plt.figure(figsize=(8,10))
plt.barh(toss_teams,toss,color="r")

plt.xlabel("No. of Toss")
plt.ylabel("Name of Teams")
plt.title("TOSS WON BY EACH TEAMS")
plt.show()
```

#**Extracting records of team won while batting first**

```python
bat_first = df[df['win_by_runs'] != 0]
bat_first.head()
```

```python
# Making histogram
plt.figure(figsize=(12,6))
plt.hist(bat_first['win_by_runs'])
plt.title("Distribution of runs")
plt.xlabel("Runs")
plt.ylabel("Wins times")

plt.show()
```

#**Extracting records of team won while batting second**

```python
bat_second = df[df['win_by_wickets']!= 0]
```

```python
# Making histogram
plt.figure(figsize=(12,6))
plt.hist(bat_second['win_by_wickets'],bins=20)
plt.title("Distribution of wickets")
plt.xlabel("Runs")
plt.ylabel("Wins times")

plt.show()
```
